```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
name_stems = ["model_vgg", "model_fusion_v2", "model_plain_v3", "model_augmented_loss_v4"]
name_stems = [x + ".pkl" for x in name_stems]
display_names = ["VGG", "Fusion", "Encoder/Decoder", "Fusion + Altered Loss"]
```

```python
colors = ["red", "green", "blue", "black"]
```

```python
plt.gcf().subplots_adjust(bottom=0.15)
fig = plt.figure()
fig.show()
ax = fig.add_subplot(111)

num_epochs = 10
x = np.arange(num_epochs)
for i, name_stem in enumerate(name_stems):
    with open(name_stem, 'rb') as f:
        dat = pickle.load(f)
        ax.plot(x, dat["loss"], color=colors[i], label=display_names[i])
    
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1),
          fancybox=True, shadow=True, ncol=2)
ax.set_title("Training Loss")

plt.savefig("training_losses.png")
```

```python
plt.tight_layout()
fig = plt.figure()
fig.show()
ax = fig.add_subplot(111)

num_epochs = 10
x = np.arange(num_epochs)
for i, name_stem in enumerate(name_stems):
    with open(name_stem, 'rb') as f:
        dat = pickle.load(f)
        ax.plot(x, dat["val_loss"], color=colors[i], label=display_names[i])
    
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1),
          fancybox=True, shadow=True, ncol=2)
ax.set_title("Validation Loss")

# plt.show()
# plt.draw()
plt.savefig("validation_losses.png")
```
